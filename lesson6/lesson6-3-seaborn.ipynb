```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
#matplotlib.style.use('seaborn')
sns.set(rc={'figure.figsize': (10, 6)})
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
ax = df[df.species == 'setosa'].plot.scatter(
    x='sepal_length', y='sepal_width', label='setosa', c='steelblue',
)
df[df.species == 'versicolor'].plot.scatter(
    x='sepal_length', y='sepal_width', label='versicolor', c='orange', ax=ax
)
df[df.species == 'virginica'].plot.scatter(
    x='sepal_length', y='sepal_width', label='virginica', c='green', ax=ax
)
```

```python
scatter_matrix(df, figsize=(10, 10))
```

```python
sns.scatterplot("sepal_length", "sepal_width", data=df, hue="species")
```

```python
sns.pairplot(df, hue='species')
```

```python
sns.jointplot("petal_width", "petal_length", data=df, kind="reg")
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.tip.plot(kind='hist')
```

```python
variables = ['sex', 'smoker', 'day', 'time', 'size']
```

```python
fig, axes = plt.subplots(nrows=len((variables)))
fig.set_figheight(20)

for i, v in enumerate(variables):
    sns.scatterplot(tips.total_bill, tips.tip, hue=tips[v], ax=axes[i])
```
